Regular Expressions allow us to search for general patterns in text data instead of searching for the exact word or phrase.

Eg: for an email we are looking for 'text'+**@**+'text'+**".com"**

We use **re** for regular expressions. This module allows us to create specialized pattern strings and then search for matches within text

Actual string: (555)-555-5555

Regex patterns: r"(\d\d\d)-\d\d\d-\d\d\d\d" --> placeholders

**SEARCHING FOR SPECIFIC STRINGS USING REGEX**

```python
import re
```

```python
text="The agent's phone number is 9444881289, phone 2"
pattern = "phone"
```

re.search( )

```python
#this will give us more info than "phone in text"
#it will only find the first match, no repeats
match = re.search(pattern,text)
match
```

```python
match.start()
```

```python
match.end()
```

re.findall( )

```python
#to find all of the matches we use
matches = re.findall(pattern,text)
matches
```

re.finditer( )

```python
#for iteratiing throught the text object and finding all of the mathches
for match in re.finditer('phone',text):
    print(match.span())
    print(match.group())
```

**BUIDLING PATTERNS USING INDETIFIER SYNTAX**

```python
text = "My phone number is 408-555-1234"
phone = re.search(r'\d\d\d-\d\d\d-\d\d\d\d',text)
print(phone)
```

```python
phone2 = re.search(r'\d{3}-\d{3}-\d{4}',text)
print(phone2)
```

```python
phone_pattern = re.compile(r"(\d{3})-(\d{3})-(\d{4})")
```

```python
results = re.search(phone_pattern,text)
```

```python
results.group()
```

```python
#indexing starts at 1
results.group(1)
```

```python
results.group(2)
```

```python
#using the or operator (|)
re.search(r'cat|dog',"the cat is here")
```

```python
#using wildcard(. --> one letter before text)
re.findall(r'.at',"The cat in the hat went splat")
```

```python
#starts with (^)
re.findall(r"^\d","1 is a number")
```

```python
#ends with ($)
re.findall(r"\d$","this is 1")
```

```python
#group to exclude something (put it in [^])
pattern = '[^\d]+'
re.findall(pattern,"This is 1 sentence with 2 numbers")
```

```python
#group to include something (put it in)
text = "Only find hyphen-words when you don't know how long-ish they are"
pattern = r'[\w]+-[\w]+' #group of alpha-nums, then a "-" then more alpha-nums that occur more than once
```

```python
re.findall(pattern,text)
```
